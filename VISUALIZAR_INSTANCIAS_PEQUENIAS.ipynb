```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Definimos los parámetros para la visualización
D = 5  # Días de la semana (lunes a viernes)
T = 7  # Bloques horarios por día

def visualizar_horario(resultados):
    """
    Visualiza el horario basado en los resultados dados para cada asignatura, sala, día y bloque.

    Parámetros:
    resultados (dict): Diccionario donde cada clave es una tupla (asignatura, sala, día, bloque)
                       y el valor es 1 si la asignatura está asignada a ese bloque.
    """
    # Crear un DataFrame vacío para los bloques y días, con índice de "Bloque 0" a "Bloque 6"
    horario_df = pd.DataFrame('', index=[f'Bloque {i}' for i in range(T)], columns=['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'])

    # Llenar el DataFrame basado en los resultados
    for (i, s, d, t), valor in resultados.items():
        if valor == 1:
            # Convertir día (d) a nombre de columna y bloque (t) a fila en el DataFrame
            dia = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'][d]
            horario_df.at[f'Bloque {t}', dia] = i  # Asignar el número de la asignatura

    # Visualización de la tabla
    fig, ax = plt.subplots(figsize=(8, 4))

    # Ocultar el eje y crear la tabla
    ax.axis('off')
    tabla = ax.table(cellText=horario_df.values, colLabels=horario_df.columns, rowLabels=horario_df.index, cellLoc='center', loc='center')

    # Personalizar el estilo de la tabla
    tabla.auto_set_font_size(False)
    tabla.set_fontsize(10)
    tabla.scale(1.2, 1.2)

    # Colorear celdas con asignaciones en un color específico
    for (i, j), cell in tabla.get_celld().items():
        if (i > 0 and j > -1) and horario_df.iat[i-1, j] != '':  # Omitir header
            cell.set_facecolor("#FFA07A")  # Color salmón para las celdas asignadas

    plt.show()

```
